###**ðŸ§  Introduction to CFG**

A **context-free grammar (CFG)** is a set of recursive rewriting rules (or productions) used to generate patterns of strings. It consists of:

* **Non-terminals (N)**: symbols that can be expanded (e.g. S, NP, VP)
* **Terminals (Î£)**: actual words in the sentence
* **Start symbol (S)**
* **Production rules (R)**: rules defining how symbols can be rewritten

Example rule:

> *S* â†’ NP VP

> NP â†’ Det N

> VP â†’ V NP

> Det â†’ '*the*' | '*a*'

> N â†’ '*cat*' | '*dog*'

> V â†’ '*chases*' | '*sees*'


###**ðŸ’» Step 1: Setup Environment**

````python
# Install NLTK (if not already installed)
!pip install nltk

# Import necessary modules
import nltk
from nltk import CFG
from nltk.parse import ChartParser
````

---

###**ðŸ§© Step 2: Example 1 â€” Simple CFG and Parse Tree**

````python
# Define a simple grammar
grammar = CFG.fromstring("""
S -> NP VP
NP -> Det N
VP -> V NP
Det -> 'the' | 'a'
N -> 'cat' | 'dog'
V -> 'chases' | 'sees'
""")

# Create a parser
parser = ChartParser(grammar)

# Define a sentence
sentence = ['the', 'cat', 'chases', 'a', 'dog']

# Parse the sentence
for tree in parser.parse(sentence):
    print(tree)
    tree.pretty_print()
````

This code defines a basic grammar and uses NLTKâ€™s ChartParser to parse and visualize the sentence `the cat chases a dog`.

---

###**ðŸ§® Step 3: Example 2 â€” Extend the Grammar with Adjectives and Prepositions**

Example structure:

```
S -> NP VP
NP -> Det Adj N | Det N
VP -> V NP PP | V NP
PP -> P NP
Adj -> 'small' | 'lazy'
P -> 'on' | 'with'
```

Try this:

````python
# Extended grammar example
grammar2 = CFG.fromstring("""
S -> NP VP
NP -> Det N | Det Adj N
VP -> V NP | V NP PP | V PP
PP -> P NP
Det -> 'the' | 'a'
N -> 'dog' | 'cat' | 'mat'
V -> 'sits' | 'chases'
Adj -> 'lazy' | 'small'
P -> 'on' | 'with'
""")

parser2 = ChartParser(grammar2)

# Example sentence
sentence2 = ['the', 'lazy', 'dog', 'sits', 'on', 'the', 'mat']

for tree in parser2.parse(sentence2):
    print(tree)
    tree.pretty_print()
````
