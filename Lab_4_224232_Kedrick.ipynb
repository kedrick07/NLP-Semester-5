###**ðŸ§  Introduction to CFG**

A **context-free grammar (CFG)** is a set of recursive rewriting rules (or productions) used to generate patterns of strings. It consists of:

* **Non-terminals (N)**: symbols that can be expanded (e.g. S, NP, VP)
* **Terminals (Î£)**: actual words in the sentence
* **Start symbol (S)**
* **Production rules (R)**: rules defining how symbols can be rewritten

Example rule:

> *S* â†’ NP VP

> NP â†’ Det N

> VP â†’ V NP

> Det â†’ '*the*' | '*a*'

> N â†’ '*cat*' | '*dog*'

> V â†’ '*chases*' | '*sees*'


###**ðŸ’» Step 1: Setup Environment**

````python
# Install NLTK (if not already installed)
!pip install nltk

# Import necessary modules
import nltk
from nltk import CFG
from nltk.parse import ChartParser
````

---

###**ðŸ§© Step 2: Example 1 â€” Simple CFG and Parse Tree**

````python
# Define a simple grammar
grammar = CFG.fromstring("""
S -> NP VP
NP -> Det N
VP -> V NP
Det -> 'the' | 'a'
N -> 'cat' | 'dog'
V -> 'chases' | 'sees'
""")

# Create a parser
parser = ChartParser(grammar)

# Define a sentence
sentence = ['the', 'cat', 'chases', 'a', 'dog']

# Parse the sentence
for tree in parser.parse(sentence):
    print(tree)
    tree.pretty_print()
````

This code defines a basic grammar and uses NLTKâ€™s ChartParser to parse and visualize the sentence `the cat chases a dog`.

###**ðŸ§® Step 3: Example 2 â€” Extend the Grammar with Adjectives and Prepositions**

Example structure:

```
S -> NP VP
NP -> Det Adj N | Det N
VP -> V NP PP | V NP
PP -> P NP
Adj -> 'small' | 'lazy'
P -> 'on' | 'with'
```

Try this:

````python
# Extended grammar example
grammar2 = CFG.fromstring("""
S -> NP VP
NP -> Det N | Det Adj N
VP -> V NP | V NP PP | V PP
PP -> P NP
Det -> 'the' | 'a'
N -> 'dog' | 'cat' | 'mat'
V -> 'sits' | 'chases'
Adj -> 'lazy' | 'small'
P -> 'on' | 'with'
""")

parser2 = ChartParser(grammar2)

# Example sentence
sentence2 = ['the', 'lazy', 'dog', 'sits', 'on', 'the', 'mat']

for tree in parser2.parse(sentence2):
    print(tree)
    tree.pretty_print()
````

**Input:** `mysentences.txt`

**Output:** Tagged tokens and parse trees in `output_trees.txt`


````python
import os
import re
import string
import io
import sys
import nltk
from nltk import CFG, ChartParser, word_tokenize

nltk.download('punkt_tab', quiet=True)

repo_url = 'https://github.com/kedrick07/NLP-Semester-5.git'
repo_dir = 'NLP-Semester-5'
if not os.path.isdir(repo_dir):
    os.system(f'git clone {repo_url}')

target_dir = None
for root, dirs, filelist in os.walk(repo_dir):
    if 'mysentences.txt' in filelist:
        target_dir = root
        break
if target_dir is None:
    raise FileNotFoundError("mysentences.txt not found in repo.")
os.chdir(target_dir)

with open('mysentences.txt', 'r') as f:
    raw_sentences = [line.strip() for line in f if line.strip()]

cleaned_sentences = []
for s in raw_sentences:
    s_clean = re.sub(r'^\s*\d+\s*[\.\)]\s*', '', s).strip()
    if s_clean:
        cleaned_sentences.append(s_clean)

grammar_str = '''
S -> NP VP | NP VP SBAR
SBAR -> Conj NP VP
NP -> Det N | Det Adj N | Det N PP | N
VP -> V NP | V PP | V NP PP | V Adv | V NP Conj V Adv | V | V PP SBAR
PP -> P NP | P Det N
Det -> 'the' | 'a' | 'an'
Adj -> 'young' | 'black' | 'red'
N -> 'student' | 'book' | 'library' | 'dog' | 'ball' | 'teacher' | 'board' | 'students'
V -> 'reads' | 'chases' | 'barks' | 'writes' | 'arrive'
P -> 'in' | 'on'
Adv -> 'loudly'
Conj -> 'and' | 'when'
'''

grammar = CFG.fromstring(grammar_str)
parser = ChartParser(grammar)

output_file = 'output_trees.txt'

print("="*70)
with open(output_file, 'w') as out:
    for idx, sentence in enumerate(cleaned_sentences, 1):
        tokens = word_tokenize(sentence.lower(), language='english')
        tokens = [t for t in tokens if t not in string.punctuation]

        header = f"\nSentence {idx}: {sentence}"
        separator = "="*70

        out.write(f"{separator}\n{header}\n{separator}\n")
        out.write(f"Tokens: {tokens}\n\n")

        print(header)
        print(separator)
        print(f"Tokens: {tokens}\n")

        found_tree = False
        tree_count = 0
        for tree in parser.parse(tokens):
            found_tree = True
            tree_count += 1
            pos = tree.pos()

            out.write(f"Parse Tree {tree_count}:\n")
            out.write(f"Tagged: {pos}\n\n")
            out.write(f"Bracketed notation:\n{tree}\n\n")

            old_stdout = sys.stdout
            sys.stdout = buffer = io.StringIO()
            tree.pretty_print()
            sys.stdout = old_stdout
            tree_diagram = buffer.getvalue()

            out.write(f"Tree diagram:\n{tree_diagram}\n")

            print(f"Parse Tree {tree_count}:")
            print(f"Tagged: {pos}\n")
            print(f"Bracketed notation:\n{tree}\n")
            print(f"Tree diagram:")
            tree.pretty_print()
            print()

        if not found_tree:
            msg = "ERROR: No parse tree found!"
            out.write(f"{msg}\n\n")
            print(f"{msg}\n")

print("="*70)
print(f"Output saved to: {output_file}")
print("="*70)

download = input("\nDo you want to download the output file? (yes/no): ").strip().lower()
if download in ['yes', 'y']:
    from google.colab import files
    files.download(output_file)
    print(f"\nâœ“ {output_file} downloaded successfully!")
else:
    print(f"\nâœ“ {output_file} is saved in your working directory.")
````
